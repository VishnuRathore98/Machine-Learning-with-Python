### Logistic Regression

#### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
# print(y)
```

```python
dataset.head(5)
```

#### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train_std = sc_X.fit_transform(X_train)
X_test_std =  sc_X.fit_transform(X_test)

# We don't need to standardize y values as that are already in the range of 0 and 1.
# sc_y = StandardScaler()
# y_train_std = sc_y.fit_transform(y_train.reshape(-1,1))
# y_test_std =  sc_y.fit_transform(y_test.reshape(-1,1))

# print("X train standard: ", X_train_std)
# print("X test standard: ", X_test_std)
```

#### Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train_std,y_train)
```

#### Predicting a new result

```python
y_test[7]
```

```python
y_predicted = classifier.predict(X_test_std)
```

```python
# print(np.column_stack((y_test,y_predicted)))
matching_count = np.sum(y_test==y_predicted)
percentage_matching = (matching_count / len(y_test))*100
print(percentage_matching)
```

```python
from sklearn.metrics import r2_score, accuracy_score, confusion_matrix, classification_report
# r2_score(y_true=y_test,y_pred=y_predicted)
# accuracy_score(y_true=y_test,y_pred=y_predicted)*100
# confusion_matrix(y_true=y_test,y_pred=y_predicted)
print(classification_report(y_test,y_predicted))
```

#### Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
